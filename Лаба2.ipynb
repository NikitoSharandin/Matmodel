```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from scipy.integrate import odeint 
data = { 2003: 504400,
      2005: 502800,
      2009: 502535,
      2010: 502015,
      2012: 508066,
      2013: 509098,
      2014: 509719,
      2015: 510152,
      2016: 510020,
      2017: 510439,
      2018: 509735,
      2019: 509420,
      2020: 508573,
      2021: 503216,
      2022: 496447

}
years=data.keys()
people=data.values()
years=np.array(list(years))
people=np.array(list(people))

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
    k=0.16
    return k * x

solve = odeint(dydt, y0=55000, t=years)
solve = solve.flatten()

plt.plot(years, solve, label = r"Мягкая модель Мальтуса")
plt.plot(years, people, color="red", label = r"Население Липецка")
plt.grid(True)
plt.legend()
plt.show()
```
