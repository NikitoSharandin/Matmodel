```python
#Номер 1
## Шарандин Никита
## x'=kx
### Жесткая модель Мальтуса
### Скорость роста пропорциональна количеству катастроф. 
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
    k=0.467
    return k*x

t=np.linspace(0, 10, 100)
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"x'=kx")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Номер 2
## Шарандин Никита
## x'=k(x)x, k(x)=a-bx
### Мягкая модель Мальтуса
### k(x) используется при слишком больших x, когда конкуренция за ресурсы приводит к уменьшению k
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

a=4
b=18
def k(x):
    return a-b*x
def dydt(x,t):
    return k(x)*x


t=np.linspace(0, 10, 100)
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"x'=ax-bx^2")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Номер 3
## Шарандин Никита
## x'=x-x^2-c
### Жесткая модель рыболовства
### При росте интенсивности рыболовства падает популяция рыбы
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
    c=0.006
    return x-x*x-c

t=np.linspace(0, 10, 100)
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"x'=x-x^2-C")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Номер 4
## Шарандин Никита
## x'=x-g(x)x-c
### Мягкая модель рыболовста
### g(x) используется при слишком больших x, чтобы увелить рост популяции при рыболовстве
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

a=0.0000004
b=0.00000018
def g(x):
    return a-b*x
def dydt(x,t):
    c=0.0000006
    return x-g(x)*x-c

t=np.linspace(0, 10, 100)
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"x'=x-g(x)*x-c")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Номер 5
## Шарандин Никита
## x'=x-x^2-kx
### Мягкая модель c дифференциальной квотой
### k позволяет узнать падение популяции
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dydt(x,t):
    k=0.006
    return x-x*x-k*x

t=np.linspace(0, 10, 100)
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
%matplotlib inline
plt.plot(t, solve, label=r"x'=x-x^2-k*x")
plt.grid(True)
plt.legend()
plt.show()
```
